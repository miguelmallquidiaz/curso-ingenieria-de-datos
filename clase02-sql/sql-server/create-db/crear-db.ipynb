## Crear una base de datos en SQL Server

- Para crear se debe tener 5 años anteriores para ver el tamaño que va a tener 2019, 2020, 2021, 2022, 2023, 2024.
- La base de datos se contruye por páginas como si fuera un libro y cada página ocupa 8 KByte en el disco.
El log debe ser la tercera parte o el 25% de la base de datos original no debe pasar de ese tamaño.
- Recordar que los archivos deben estar en otras unidades y no en la unidad C es caso de ataques.
