## Resumen

| Categoría | Tipo          | Comandos principales                                      |
| --------- | ------------- | --------------------------------------------------------- |
| **DDL**   | Definición    | `CREATE`, `ALTER`, `DROP`, `TRUNCATE`                     |
| **DML**   | Manipulación  | `INSERT`, `UPDATE`, `DELETE` *(a veces también `SELECT`)* |
| **DQL**   | Consulta      | `SELECT`                                                  |
| **DCL**   | Control       | `GRANT`, `REVOKE`                                         |
| **TCL**   | Transacciones | `COMMIT`, `ROLLBACK`, `SAVE TRANSACTION`                  |
