## Uso de Databricks Community Edition

- En Compute tiene que estar encendido, si no esta darle al icono de play.



- Para estructurar mejor tus archivos puedes crear un folder



- Para importar el archivo ipynb le tienes que dar al icono de los tres puntos.



- Subes el archivo ipynb



- Solamente creas el nombre del nuevo schema y el volumen



- Para subir los archivos hay que crear un nuevo volumen



En databricks es posible tener el dataframe como si fuera una hoja en excel, hay distintas formas.
- order_details.display()
- display(order_details)
- order_details.show()
