## Aprendiendo Excel

### Validación de datos

La validación de datos en Excel es una herramienta que te permite controlar lo que los usuarios pueden escribir en una celda o rango de celdas.

Pasos a seguir:

1. Seleccionar la celda en que vas a validar.
2. Ingresar a la pestaña Datos > Herramientas de datos > validación de datos.
3. Poner los criterios a validar.



### Filtros avanzados

- Los filtros avanzados en Excel permiten extraer y mostrar registros específicos de una tabla o base de datos según criterios más complejos o múltiples condiciones.

Pasos a seguir:
1. Debes tener los criterios que quieres filtrar.
2. Te posicionas en una celda.
3. Ahora ir a la pestaña Datos > Avanzadas.
4. Habilitar la opción copiar en otro lugar.
5. Seleccionas la tabla empresas, luego los criterios con encabezados y todo, al final seleccionas el lugar en donde se va a copiar.
6. Como consejo es mejor eliminar el rango o verificar que copie bien.

| Función              | Tipo de búsqueda                 | Devuelve                   | Sintaxis básica                                                     | Ventajas principales                        |
| -------------------- | -------------------------------- | -------------------------- | ------------------------------------------------------------------- | ------------------------------------------- |
| **BUSCARV**          | Vertical (hacia abajo)           | Valor a la **derecha**     | `=BUSCARV(valor, rango, nro_columna, [coincidencia])`               | Fácil, muy usada; ideal para tablas simples |
| **BUSCARH**          | Horizontal (hacia la derecha)    | Valor en **fila inferior** | `=BUSCARH(valor, rango, nro_fila, [coincidencia])`                  | Útil si los encabezados están en filas      |
| **XLOOKUP**          | Flexible (arriba/abajo/izq/dcha) | Valor relacionado          | `=XLOOKUP(valor_buscado, rango_buscar_en, rango_devolver)`          | Más moderna; no necesita contar columnas    |
| **COINCIDIR**        | Busca posición                   | Número (posición)          | `=COINCIDIR(valor, rango, [tipo])`                                  | Devuelve la posición del dato               |
| **INDICE**           | Devuelve por posición            | Valor                      | `=INDICE(rango, núm_fila, [núm_columna])`                           | Muy flexible cuando se combina              |
| **INDICE+COINCIDIR** | Búsqueda avanzada                | Valor                      | `=INDICE(rango_resultado, COINCIDIR(valor_clave, rango_buscar, 0))` | Reemplaza al BUSCARV con más control        |
